```python
import random

def create_board():
    return [[" "]*3 for _ in range(3)]

def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 9)

def check_winner(board, player):
    win_positions = [
        [(0, 0), (1, 0), (2, 0)], [(0, 1), (1, 1), (2, 1)], [(0, 2), (1, 2), (2, 2)],  # colunas
        [(0, 0), (0, 1), (0, 2)], [(1, 0), (1, 1), (1, 2)], [(2, 0), (2, 1), (2, 2)],  # linhas
        [(0, 0), (1, 1), (2, 2)], [(0, 2), (1, 1), (2, 0)]  # diagonais
    ]

    for positions in win_positions:
        symbols = [board[i][j] for i, j in positions]
        if all(symbol == player for symbol in symbols):
            return True, positions
    return False, None

def play_game():
    board = create_board()
    players = ['X', 'O']
    current_player = random.choice(players)
    print(f"O jogador '{current_player}' foi escolhido para começar.")

    for turn in range(9):
        print_board(board)
        print(f"Vez do jogador '{current_player}' ({turn+1}/9)")

        while True:
            try:
                row = int(input("Digite a linha (0, 1 ou 2): "))
                col = int(input("Digite a coluna (0, 1 ou 2): "))
                if board[row][col] == " ":
                    board[row][col] = current_player
                    break
                else:
                    print("Essa posição já está preenchida. Tente novamente.")
            except (ValueError, IndexError):
                print("Entrada inválida. Digite um número entre 0 e 2.")

        if turn >= 4:
            winner, positions = check_winner(board, current_player)
            if winner:
                print_board(board)
                print(f"O jogador '{current_player}' venceu nas posições: {positions}")
                return

        current_player = 'O' if current_player == 'X' else 'X'

    print_board(board)
    print("Ih, o jogo deu Nova.")

while True:
    play_game()
    replay = input("Deseja jogar novamente? (s/n): ").lower()
    if replay != 's':
        break

```
